```python
students = {}

while True:
    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Show Topper")
    print("7. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        roll = input("Enter Roll Number: ")
        name = input("Enter Name: ")
        student_class = input("Enter Class: ")

        math = int(input("Math Marks: "))
        science = int(input("Science Marks: "))
        english = int(input("English Marks: "))

        total = math + science + english
        percentage = total / 3

        # Grade Calculation
        if percentage >= 90:
            grade = "A+"
        elif percentage >= 80:
            grade = "A"
        elif percentage >= 70:
            grade = "B"
        elif percentage >= 60:
            grade = "C"
        elif percentage >= 50:
            grade = "D"
        else:
            grade = "F"

        students[roll] = {
            "name": name,
            "class": student_class,
            "math": math,
            "science": science,
            "english": english,
            "total": total,
            "percentage": percentage,
            "grade": grade
        }

        print("Student Added Successfully!")

    elif choice == "2":
        if not students:
            print("No Student Records Found!")
        else:
            for roll, data in students.items():
                print("\n-------------------")
                print("Roll No:", roll)
                print("Name:", data["name"])
                print("Class:", data["class"])
                print("Math:", data["math"])
                print("Science:", data["science"])
                print("English:", data["english"])
                print("Total:", data["total"])
                print("Percentage:", round(data["percentage"], 2))
                print("Grade:", data["grade"])

    elif choice == "3":
        roll = input("Enter Roll Number: ")

        if roll in students:
            data = students[roll]
            print("\nStudent Details")
            print("Name:", data["name"])
            print("Class:", data["class"])
            print("Math:", data["math"])
            print("Science:", data["science"])
            print("English:", data["english"])
            print("Total:", data["total"])
            print("Percentage:", round(data["percentage"], 2))
            print("Grade:", data["grade"])
        else:
            print("Student Not Found!")

    elif choice == "4":
        roll = input("Enter Roll Number: ")

        if roll in students:
            new_name = input("Enter New Name: ")
            students[roll]["name"] = new_name
            print("Record Updated!")
        else:
            print("Student Not Found!")

    elif choice == "5":
        roll = input("Enter Roll Number: ")

        if roll in students:
            del students[roll]
            print("Student Deleted!")
        else:
            print("Student Not Found!")

    elif choice == "6":
        if students:
            topper = max(
                students,
                key=lambda x: students[x]["percentage"]
            )

            print("\n===== Topper Details =====")
            print("Roll No:", topper)
            print("Name:", students[topper]["name"])
            print("Percentage:", round(students[topper]["percentage"], 2))
            print("Grade:", students[topper]["grade"])
        else:
            print("No Student Records Found!")

    elif choice == "7":
        print("Program Ended.")
        break

    else:
        print("Invalid Choice!")
```
